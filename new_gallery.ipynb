```python
%%capture_png -p Header1.png --celltype Header

import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import ImageGrid
import matplotlib
import numpy as np

def set_circle(array, radius, value):
    x, y = np.indices(array.shape)
    new_array = array.copy()
    new_array[(x-int(len(x)/2))**2 + (y-int(len(y)/2))**2 < radius**2] = value
    return new_array

array = np.zeros((101, 101))
im1 = set_circle(array, 3*12, 1)
im2 = set_circle(im1, 3*10, 2)
im3 = set_circle(im2, 3*8, 3)
im4 = set_circle(im3, 3*6, 4)
plt.rcParams['figure.dpi'] = 300


import matplotlib.pyplot as plt #NOT
fix, ax = plt.subplots(figsize=(5,5), facecolor="#F7F5F3") #NOT
text_kwargs = dict(ha="center", va="center", fontsize=70, color="#ED6A5A") #NOT
ax.text(0.5, 0.5, "Header",**text_kwargs) #NOT
ax.axis("off"); #NOT
```

```python
%%capture_png -p Normal.png  --celltype Normal --style "height: 150px; width: auto;"
fig = plt.figure(figsize=(4, 4))
grid = ImageGrid(fig, 111,
                 nrows_ncols=(2, 2),
                 axes_pad=0.3,
                 label_mode="1",
                 cbar_size="7%",
                 cbar_pad="0%",
                 )

for ax, img, cax in zip(grid, [im1, im2, im3, im4], grid.cbar_axes):
    im = ax.imshow(img, cmap="YlGnBu")
```

```python
%%capture_png -p LabelMode.png  --celltype Normal --style  "height: 75px; width: auto;"
from matplotlib.offsetbox import AnchoredText

fig = plt.figure(figsize=(4, 4))
grid = ImageGrid(fig, 111,
                  nrows_ncols=(1, 4),
                  axes_pad=0.05,
                  label_mode="1",
                  share_all=False,
                  cbar_location="right",
                  cbar_mode="single",
                  cbar_size="10%",
                  cbar_pad=0.05,
                  )
vmax, vmin = np.max(im4), np.min(im4)
norm = matplotlib.colors.Normalize(vmax=vmax, vmin=vmin)
labels = ["a)", "b)", "c)", "d)"]
for ax, img, cax, label in zip(grid, [im1, im2, im3, im4], grid.cbar_axes , labels):
    im = ax.imshow(img, cmap="YlGnBu", norm=norm)
    at = AnchoredText(label,loc="lower left", pad=0., borderpad=0.5, frameon=False)
    ax.add_artist(at)
    ax.axis("off")
ax.cax.colorbar(im)
```

```python
%%capture_png -p LabelMode.png  --celltype Normal --style  "height: 75px; width: auto;"
from matplotlib.offsetbox import AnchoredText

fig = plt.figure(figsize=(4, 4))
grid = ImageGrid(fig, 111,
                  nrows_ncols=(1, 4),
                  axes_pad=0.05,
                  label_mode="1",
                  share_all=False,
                  cbar_location="right",
                  cbar_mode="single",
                  cbar_size="10%",
                  cbar_pad=0.05,
                  )
vmax, vmin = np.max(im4), np.min(im4)
norm = matplotlib.colors.Normalize(vmax=vmax, vmin=vmin)
labels = ["a)", "b)", "c)", "d)"]
for ax, img, cax, label in zip(grid, [im1, im2, im3, im4], grid.cbar_axes , labels):
    im = ax.imshow(img, cmap="YlGnBu", norm=norm)
    at = AnchoredText(label, prop=dict(size=8), frameon=True, loc='upper left')
    at.patch.set_boxstyle("round,pad=0.,rounding_size=0.0")
    ax.add_artist(at)
    ax.axis("off")

ax.cax.colorbar(im)


```
