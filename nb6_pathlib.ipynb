```python
import matplotlib.pyplot as plt
from matplotlib.transforms import Affine2D

def rainbow_text(x, y, strings, colors, orientation='horizontal',
                 ax=None, **kwargs):
    if ax is None:
        ax = plt.gca()
    t = ax.transData
    canvas = ax.figure.canvas
    for s, c in zip(strings, colors):
        text = ax.text(x, y, s , color=c, transform=t, **kwargs)

        # Need to draw to update the text position.
        text.draw(canvas.get_renderer())
        ex = text.get_window_extent()
        if orientation == 'horizontal':
            t = text.get_transform() + Affine2D().translate(ex.width, 0)

words = str(mypath)
words = words.replace("/", "/ ")
words = words.replace(".", " . ")
words = words.split()
print(words)
colors = ['red', 'orange', 'gold', 'lawngreen', 'lightseagreen', 'royalblue', # NOT
          'blueviolet', "red"] 
fix, ax = plt.subplots(figsize=(18, 1), facecolor="#F7F5F3") 
rainbow_text(0.01, 0.2, words, colors, size=40)
ax.axis("off"); 

def plot_text(words, colors):
    fix, ax = plt.subplots(figsize=(18, 1), facecolor="#F7F5F3")  
    rainbow_text(0.01, 0.2, words, colors, size=40) 
    ax.axis("off");  
```

```python
from PIL import Image
import requests
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl

from PIL import Image
import numpy as np
import urllib.request
from pathlib import Path
```

```python
ChapterManager.set_chapter_name("Pathlib -  Further Functions")

#get apple emoji
def get_apple_emoji(emoji):
    emoji_code = "-".join(f"{ord(c):x}" for c in emoji)
    url = f"https://raw.githubusercontent.com/cj1128/emoji-images/master/imgs/{emoji_code}.png"
    im = Image.open(requests.get(url, stream=True).raw)
    #im = np.array(im.convert("RGBA"))
    return im


def plot_apple_emojis(e1, e2, overlap =0.5):
    fig, (ax1, ax2) = plt.subplots(1, 2)
    im_e1 = get_apple_emoji(e1)
    im_e2 = get_apple_emoji(e2)

    ax1.imshow(im_e1)
    ax1.axis("off")
    ax2.imshow(im_e2)
    ax2.axis("off")
    plt.subplots_adjust(wspace=-overlap)
```
