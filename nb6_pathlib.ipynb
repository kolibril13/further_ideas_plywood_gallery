```python
from PIL import Image
import requests
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl
mpl.rcParams['axes.axisbelow'] = False
def get_emoji(emoji):

    emoji_code = "-".join(f"{ord(c):x}" for c in emoji)
    url = f"https://raw.githubusercontent.com/cj1128/emoji-images/master/imgs/{emoji_code}.png"
    im = Image.open(requests.get(url, stream=True).raw)
    #im = np.array(im.convert("RGBA"))
    return im


def plot_emojis(e1, e2, overlap =0.5):
    fig, (ax1, ax2) = plt.subplots(1, 2)
    im_e1 = get_emoji(e1)
    im_e2 = get_emoji(e2)

    ax1.imshow(im_e1)
    ax1.axis("off")
    ax2.imshow(im_e2)
    ax2.axis("off")
    plt.subplots_adjust(wspace=-overlap)
        
```

```python

from PIL import Image
import numpy as np
import urllib.request
from pathlib import Path

import zipfile
path =  Path.cwd() / f"temp_emojis"

if not path.exists(): # Download and unzip assets
    path_zip  = Path.cwd() / f"openmoji-618x618-color.zip"
    url = f"https://github.com/hfg-gmuend/openmoji/releases/download/13.1.0/openmoji-618x618-color.zip"
    urllib.request.urlretrieve(url, path_zip)
    with zipfile.ZipFile(path_zip,"r") as zip_ref:
        zip_ref.extractall("temp_emojis")
    path_zip.unlink()

def get_open_emoji(emoji):

    emoji_code = "-".join(f"{ord(c):x}" for c in emoji)
    im = Image.open(path / f"{emoji_code}.png")
    #im = np.array(im.convert("RGBA"))
    return im

def plot_open_emojis(e1, e2, overlap =0.5):
    fig, (ax1, ax2) = plt.subplots(1, 2)
    im_e1 = get_open_emoji(e1)
    im_e2 = get_open_emoji(e2)

    ax1.imshow(im_e1)
    ax1.axis("off")
    ax2.imshow(im_e2)
    ax2.axis("off")
    plt.subplots_adjust(wspace=-overlap)
plot_open_emojis("üìÅ","‚ùì" ) # NOT
plot_open_emojis("üìÅ","üóë", overlap=0.4) # NOT
```
