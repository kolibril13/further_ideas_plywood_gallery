```python
from PIL import Image
import requests
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl
mpl.rcParams['axes.axisbelow'] = False
def get_emoji(emoji, emoji_code_custom=False):
    if emoji:
        emoji_code = str(emoji.encode("unicode_escape"))[8:-1]  # trim
    if emoji_code_custom:
        emoji_code = emoji_code_custom
    url = f"https://raw.githubusercontent.com/cj1128/emoji-images/master/imgs/{emoji_code}.png"
    im = Image.open(requests.get(url, stream=True).raw)
   # im = np.array(im.convert("RGBA"))
    return im

def plot_emojis(e1, e2, overlap =0.5):
    fig, (ax1, ax2) = plt.subplots(1, 2)
    ax1.imshow(e1)
    ax1.axis("off")
    ax2.imshow(e2)
    ax2.axis("off")
    plt.subplots_adjust(wspace=-overlap)
        
```
